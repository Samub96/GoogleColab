2.    TIPO DE RECTIFICADORES

> * **Rectificador de media onda:** Un rectificador de media onda es un dispositivo o circuito electrónico que convierte la corriente alterna (CA) en corriente continua (CC) aprovechando solo una mitad de la onda de entrada. Este tipo de rectificador utiliza un solo diodo para permitir que la corriente fluya en una dirección durante la mitad positiva de la onda alterna.
El proceso básico de un rectificador de media onda implica bloquear una mitad de la onda de entrada y permitir el paso de la otra mitad. Durante la mitad positiva de la onda de entrada, el diodo permite que la corriente fluya a través de la carga, mientras que durante la mitad negativa, el diodo bloquea el flujo de corriente.
<img src = https://electronika2.tripod.com/info_files/pix/Image11.jpg  align="center">


>* **Rectificador de onda completa:** Un rectificador de onda completa es un dispositivo o circuito electrónico diseñado para convertir corriente alterna (CA) en corriente continua (CC) aprovechando ambas mitades de la onda alterna. Este tipo de rectificador utiliza cuatro diodos dispuestos en un puente, comúnmente conocido como el Puente de Graetz, para permitir que la corriente fluya en ambas direcciones durante cada ciclo de la onda alterna.

# Análisis teórico

##**Filtros**
Los filtros son dispositivos encargados de eliminar o atenuar ciertas frecuencias indeseadas en una señal que pasa a través de el.En el campo de la teoría electromagnetica estos sirven para controlar el rango de frecuencias que es emitido por un circuito de corriente alterna,algunas de los conceptos generales de un filtro son:

* **Banda de paso :** Rango de frecuencias en las que el filtro permite el paso,con un límite inferior $f_1$ y un límite superior $f_2$.
*   **Ancho de banda :**La diferencia entre el límite superior e inferior $Δf=f_2-f_1$
*   **Banda atenuada :**Es el rango de frecuencias par las cuales el filtro ejerce una atenuación al pasar por el.
* **frecuencia de corte :**Representada con el simbolo $f_c$ se refiere al punto donde el filtro genera una mayor o menor atenuación a la señal que pasa por él.

 Para la frecuencia de corte tenemos una formula general para los diferentes tipos de filtro:

 $$f_c=\frac{1}{2\pi RC}$$

 Donde $R$ es la resistencia  que tiene el filtro y $C$ es  la capacitancia del capacitor.

###**Filtros pasa-baja**

El más simple de este tipo de filtros se conforma por una resistencia en serie con la fuente de voltaje y un capacitor entre la resistencia y el putno de salida de voltaje para el resto del circuito y el otro teminal  conectado a tierra,esta configuración permite que las bajas frecuencias pasen por el resistor y salgan por el punto de salida de voltaje para el resto del circuito sin ser atenuadas mientras que  las frecuencias mayores a la $f_c$ una notable atenuación.Este comportamiento se explica al analizar la fórmula de redactancia del capacitor:

$$X_c=\frac{1}{2\pi fc}$$

Observamos que dicha redactancia o resistencia varía según la frecuencia de la corriente en circuito.Recordemos tambien que la resistencia es la oposición al paso de corriente, por lo que en este caso a mayor frecuencia menor resistencia hay en el capacitor por lo que la corriente empezará a ir directo al capacitor ,que tiene menor resistencia, y finalmente volverá al polo a tierra haciendo que la corriente y el voltaje a la salida del circuito sea muy pequeño y por el contrario con menor frecuencia más corriente seguirá su camino hacia la salida de voltaje suministrando mayor corriente y mayor voltaje.

###**Ganancia de potencia**
Para entender un poco mejor como los filtros ejercen una atenuación en la señal de ac que pasa por ellas,es necesario hablar de la ganancia de potencia.Pese al nombre la ganancia de potenci no solo puede indicar un aumento sino tambien una disminución que existe entre la potencia suministrada al circuito y la potencia entregada por el mismo mismo.El calculo de dichas ganancia es:

Ganancia de potencia:

$$A_p=10log_{10}\frac{P_{Sal}}{P_{Ent}}$$

Donde  $P_{Ent}$ es la potencia que se entrega a nuestro circuito,es decir,la intensidad de la señal proveniente de la fuente de AC,por otro lado, la $P_{Sal}$ es el  potencia que salen tras pasar por el circuito,en nuestro caso el filtro,el termino $10log_{10}$ hace que las unidades de las cantidades resultantes que comunmente son extremadamente pequeñas(en bels) queden en unidades más manejables los decibeles(dB).Ya que estamos hablando de filtros, esta cantidad debe ser 1 o muy cercana a este para las frecuencias que se encuentran en la región de paso del filtro, pero a partir de la $f_c$  esta cantidad debe ser cercana a los $-3dB$ o ser menor para las frecuencias  que se encuentran en la banda atenuada.

# Analisis Numerico

Vamos a platear 2 casos de analisis:

-Rectificadores de onda media


-Rectificadores de onda completa


para los cuales vamos a calcular:







**Para media onda:**

-Valor medio del Voltaje= $V_{med}=\frac{V_{Max}}{π}$

-Valor eficaz del voltaje= $V_{ef}=\frac{V_{Max}}{2}$

-Valor medio de la Corriente= $I_{med}=\frac{V_{Med}}{R}$

-Valor eficaz del voltaje= $I_{ef}=\frac{V_{ef}}{R}$

**Para onda completa:**


-Valor medio del Voltaje= $V_{med}=\frac{2V_{Max}}{π}$

-Valor eficaz del voltaje= $V_{ef}=\frac{V_{Max}}{\sqrt{2}}$

-Valor medio de la Corriente= $I_{med}=\frac{V_{Med}}{R}$

-Valor eficaz del voltaje= $I_{ef}=\frac{V_{ef}}{R}$

El analisis se hará, segun los siguientes circuitos

image.png
image.png

```python
import numpy as np
#Calculos para media onda
Vmax=5
frecuencia=40
r=100

Vmed=((2*Vmax)/np.pi)
Vef=(Vmax/np.sqrt(2))
Imed=(Vmed/r)
Ief=(Vef/r)
Imax = Vmax / r

print("Voltaje Medio:", Vmed)
print("Voltaje Eficiente:", Vef)
print("Corriente Media:", Imed)
print("Corriente Eficaz:", Ief)
print("Corriente Máxima:", Imax)


```

```python
import numpy as np
#Calculos para onda completa
Vmax=5
frecuencia=40
r=100

Vmed=(Vmax/np.pi)
Vef=(Vmax/2)
Imed=(Vmed/r)
Ief=(Vef/r)
Imax = Vmax / r
print("Voltaje Medio:", Vmed)
print("Voltaje Eficiente:", Vef)
print("Corriente Media:", Imed)
print("Corriente Eficaz:", Ief)
print("Corriente Máxima:", Imax)
```

Para el caso de un rectificador, pero con un filtro(un capacitor) tambien podemos calcular la impedancia, y la corriente pico a traves del capacitor

```python
import numpy as np
#Calculos para onda completa con filtro(capacitor)
Vmax=5
frecuencia=40 #En Herzios
r=100
C=0.001

Vmed=(Vmax/np.pi)
Vef=(Vmax/2)
Imed=(Vmed/r)
Ief=(Vef/r)
Imax = Vmax / r
Fc=1/(2*np.pi*r*C)
Xc = 1 / (2 * np.pi * frecuencia * C)
Z = np.sqrt(r**2 + Xc**2)
Ic_pico = Vmax / Z
print("Voltaje Medio:", Vmed)
print("Voltaje Eficiente:", Vef)
print("Corriente Media:", Imed)
print("Corriente Eficaz:", Ief)
print("Corriente Máxima:", Imax)
print("Impedancia Total:", Z)
print("Corriente de Pico a través del Capacitor:", Ic_pico)
print(Fc)
```
